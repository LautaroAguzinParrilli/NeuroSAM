```python
import pandas as pd
import os
```

```python
df= pd.read_csv("../Data/all_participants.csv")

# Create a DataFrame from the file paths
folder_path = "C:/Users/parri/OneDrive/Documentos/Beca PEFI/quasiraw/"
data = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        data.append({'ID': file.split('_desc-6apply_T1w.nii.gz')[0], 'Path': file_path})

file_paths = pd.DataFrame(data)
file_paths['ID'] = file_paths['ID'].str.replace('.nii.gz', '')
file_paths['ID'] = file_paths['ID'].str.replace('_T1w', '')
```

```python
csv_data = pd.merge(df, file_paths, on='ID')[['Path', 'Age']]
```
